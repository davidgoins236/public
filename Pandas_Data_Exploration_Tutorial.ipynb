### Cleaning our data


```python
raw.shape #count rows & cols, Question: why not raw()
raw.head() 
raw.isnull().sum() #figure out how many null values we have by column
```

```python
#replace NA with 0
df = raw.fillna(0)
#test with: df.isnull().sum()
#remove the rank variable
df = df.drop(['Rk'],axis = 1)
```

```python
#setup notebook to show all rows of the data
#pd.set_option('display.max_rows',df.shape[0]+1)
#revert to default
pd.set_option('display.max_rows',50)
df
```

```python
#df.select_dtypes(include=['object'])
#df.dtypes

#okay so it looks like the fastest way to fix all these 'object' data types is to save and reload as csv
df.to_csv("nba.csv")
#now pull it back in and check dtypes
df = pd.read_csv("nba.csv")
df.dtypes
```

```python
#who has the most points
df[df.PTS==df.PTS.max()]
#who are the top k players for x
df.nlargest(10,'STL')
```

```python
#how many player(s) score > 20 points per game
df[df.PTS>20].shape[0] #nrows() equivalent - how many players score > 20
```

```python
#who scores>20
df[df.PTS>20].loc[:,['Player','Pos','Tm','PTS']].sort_values('PTS',ascending = False) #.loc works a bit like select()
```

```python
#tutorial question: which player had the highest 3 pt fg per game
df[df['3P']==df['3P'].max()]
#which player lead assits
df.nlargest(10,'AST').sort_values('AST',ascending = False).loc[:,['Player','Pos','Tm','3P','PTS','AST']]
```

```python
#Which player had 2nd most points on the Lakers
lal=df[df['Tm']=='LAL']
lal.sort_values('PTS',ascending = False).iloc[[1]] #note: pandas starts counting at 0
```

```python
#which position scores most
position = df.groupby('Pos').PTS.describe()
position[position['count']>5].sort_values('mean',ascending=False)
```
