### Use pytrends to query google trends data by keyword and daterange

```python
#!pip install pytrends
```

```python
# import the TrendReq method from the pytrends request module
import pandas as pd
from pytrends.request import TrendReq

# execute the TrendReq method by passing the host language (hl) and timezone (tz) parameters
pytrends = TrendReq(hl='en-US', tz=360)

#create keyword list
kw_list = ["covid","covid vaccine","lockdown","pandemic"]
pytrends.build_payload(kw_list, cat=0, timeframe='2020-01-01 2022-05-20', geo='US', gprop='')
gt_o3_ytd1 = pytrends.interest_over_time()
gt_o3_ytd1
```

```python
#plot a timeseries 
gt_o3_ytd1.plot(figsize=(20,12))
```
